```python
import pandas as pd
import os

# Ruta de la carpeta que contiene los archivos Excel
carpeta_datos = "DatosVacunacion"

# Obtener la lista de archivos en la carpeta
archivos_excel = [archivo for archivo in os.listdir(carpeta_datos) if archivo.endswith('.xlsx')]

# Crear un diccionario para almacenar los dataframes
dataframes = {}

# Leer cada archivo Excel y almacenarlo en el diccionario
for archivo in archivos_excel:
    ruta_completa = os.path.join(carpeta_datos, archivo)
    nombre_df = os.path.splitext(archivo)[0]  # Usar el nombre del archivo como clave
    dataframes[nombre_df] = pd.read_excel(ruta_completa)

# Mostrar los nombres de los dataframes cargados
print("DataFrames cargados:", list(dataframes.keys()))
```

```python
for nombre_df, df in dataframes.items():
    print(f"DataFrame: {nombre_df}")
    print(df.head())
    print("--" * 40)
```
